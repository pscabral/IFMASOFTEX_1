```python
import math

def px_to_in(px, h, w):
    ppi = math.sqrt(h*h + w*w)
    print(ppi)
    return px / 95.995
```

```python
from pptx import Presentation
from pptx.util import Inches
import os

# Define px_to_in function if not already defined
def px_to_in(px, height, width):
    dpi = 96  # assuming standard screen DPI, you can adjust this value if needed
    inches_height = height / dpi
    inches_width = width / dpi
    return px / max(inches_height, inches_width)

path = "/content/"

prs = Presentation()
prs.slide_height = Inches(6.667)
prs.slide_width = Inches(6.667)

slide = prs.slides.add_slide(prs.slide_layouts[6])

# Use correct assignment for width
for i, box in enumerate(result.boxes):
    class_id = result.names[box.cls[0].item()]
    cords = box.xyxy[0].tolist()
    cord = [round(x) for x in cords]
    conf = round(box.conf[0].item(), 2)

    h = cords[3] - cords[1]
    w = cords[2] - cords[0]  # Fix the assignment here
    left_in = px_to_in(cords[0], h, w)
    top_in = px_to_in(cords[1], h, w)
    slide.shapes.add_picture(path + f"output-{i}.png", left=Inches(left_in), top=Inches(top_in))

# Uncomment the line to save the presentation
prs.save("Presentation.pptx")
```
